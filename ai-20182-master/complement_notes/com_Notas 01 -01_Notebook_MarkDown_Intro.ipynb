# Índices y potencias

Se puede representar una potencia en LaTex con un acento circunflejo (^). Si se tiene más de un caracter, estos se deben encerrar en llaves {}. $$f(x) = x^2 - x^{1 \over 10}$$

Para los índices se puede usar el guión bajo (`_`) así: $$f(x) = x_n + x_{n-1}$$

**Código:**
```markdown
$$f(x) = x^2 - x^{1 \over 10}$$
$$f(x) = x_n + x_{n-1}$$

```

# Raíces y logaritmos

Las raíces cuadradas se pueden representar en LaTex como `\sqrt`; sin embargo, para cambiar el nivel de dicha raíz se hace uso de `\sqrt[n]`, donde `n` es el nivel. $$f(x) = \sqrt[5]{2x} + \sqrt{x-2}$$

Para representar un logaritmo se utiliza `\log`así: $$\log_{2}{8} = 3$$

**Código:**
```
$$f(x) = \sqrt[5]{2x} + \sqrt{x-2}$$
$$\log_{2}{8} = 3$$
```

# Sumatorias y productorias

En LaTex, una suma se representa como `\sum\limits_{a}^{b}`, donde a y b son los límites inferior y superior, respectivamente.
$$\sum\limits_{x=1}^{\infty} {1 \over x} = 1$$

La productoria se describe mediante `\prod\limits_{a}^{b}`, donde a y b son los límites inferior y superior, respectivamente.
$$\prod\limits_{i=1}^{n} x_i = 1$$

**Código:**
```
$$\sum\limits_{x=1}^{\infty} {1 \over x} = 1$$
$$\prod\limits_{i=1}^{n} x_i = 1$$
```

# Cálculo

Los límites se representan usando `\lim\limits_{x \to a}`, en tanto `x` tiende `a`.
$$\lim\limits_{x \to 0^+} {1 \over 0} = \infty$$

Para las integrales se utiliza `\int\limits_{a}^{b}`, donde `a` y `b`, son el límite inferior y el superior respectivamente.
$$\int\limits_{-\pi}^{\pi} 2\theta \, d\theta$$


**Código:**
```
$$\lim\limits_{x \to 0^+} {1 \over 0} = \inf$$
$$\int\limits_{-\pi}^{\pi} 2\theta \, d\theta$$
```

# Definición de funciones segmentadas

Para definir una función que es calculada de diversas formas sobre un número determinado de períodos, se puede utilizar LaTex.

$\lvert x\rvert \equiv abs(x) =\left\{\begin{array}{ll}-x,  & \mbox{if } x < 0 \\{x}, & \mbox{if } x \geq 0\end{array}\right.$

**Código:**
```
$\lvert x\rvert \equiv abs(x) =
\left\{
    \begin{array}{ll}
        -x,  & \mbox{if } x < 0 \\
        {x}, & \mbox{if } x \geq 0
\end{array}
\right.$
```

$$\| x \|_{p}=\left( \sum_{i=1}^{n} \lvert x_{i} \rvert^{p} \right)^{^1/_p}=\sqrt[p]{\| x_{1} \|^{p}+\| x_{2} \|^{p}+ \cdots +\| x_{n} \|^{p}}$$

Tips:

- $\| \cdot \|$ es `\| \cdot \|`
- $\left( \cdot \right)$ es `\left( \cdot \right)`
- $\sqrt[p]{\cdot}$ es `\sqrt[p]{\cdot}`.
